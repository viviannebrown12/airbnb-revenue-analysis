```python
# Airbnb Revenue Analysis

## 01 — Data Understanding

### Objective

#Understand the structure, quality, and key variables in the New York City Airbnb datasets before performing deeper cleaning and analysis.

### Dataset

#This notebook uses the detailed NYC listings dataset from Inside Airbnb.

### Initial Questions

#- How many listings and variables are included?
#- Which columns are most relevant to pricing and revenue?
#- Which important variables contain missing values?
#- Which variables need to be converted or cleaned?

# Import pandas for working with tabular data
import pandas as pd
# Load the compressed listings dataset
# The notebook is inside the notebooks folder, so ../ moves up to the main project folder
listings = pd.read_csv("../data/raw/listings.csv.gz")
```

```python
# Display the first five rows to understand the general structure
listings.head()
# Return the number of rows and columns
listings.shape
```

```python
# Display all available column names
list(listings.columns)
```

```python
## Key Variables for Analysis
important_columns = [
    "price",
    "room_type",
    "property_type",
    "accommodates",
    "bedrooms",
    "beds",
    "bathrooms_text",
    "amenities",
    "host_is_superhost",
    "review_scores_rating",
    "number_of_reviews",
    "availability_365",
    "neighbourhood_cleansed"
]

listings[important_columns].head()
```

```python
listings[important_columns].isnull().sum()
```

```python
listings[important_columns].dtypes
```

```python
listings["price"].head(20)
```

```python
# Convert price values into numbers
# 1. Convert all values to pandas' string type
# 2. Remove dollar signs
# 3. Remove commas from values such as $1,250.00
# 4. Convert the cleaned text into numeric values
# 5. Turn invalid or missing values into NA rather than causing an error
listings["price_clean"] = pd.to_numeric(
    listings["price"]
    .astype("string")
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False),
    errors="coerce"
)
```

```python
# Compare the original and cleaned price columns
listings[["price", "price_clean"]].head(10)
```

```python
# Confirm that price_clean is numeric
listings["price_clean"].dtype
```

```python
# Count how many listings have a usable numeric price

listings["price_clean"].notna().sum()
```

```python
# Summary statistics for cleaned prices

listings["price_clean"].describe().round(2)
```

```python
# Compare average prices across room types

room_type_summary = (
    listings.groupby("room_type")["price_clean"]
    .agg(
        listing_count="count",
        average_price="mean",
        median_price="median"
    )
    .round(2)
    .sort_values("average_price", ascending=False)
)

room_type_summary
```

```python
# Compare average prices across property types

property_summary = (
    listings.groupby("property_type")["price_clean"]
    .agg(
        listing_count="count",
        average_price="mean",
        median_price="median"
    )
    .round(2)
    .sort_values("average_price", ascending=False)
)

property_summary.head(15)
```

```python
# Display the 10 most expensive Airbnb listings

listings[
    [
        "name",
        "room_type",
        "property_type",
        "neighbourhood_cleansed",
        "price_clean"
    ]
].sort_values(
    "price_clean",
    ascending=False
).head(10)
```
