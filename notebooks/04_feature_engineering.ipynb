```python
import pandas as pd
import numpy as np
```

```python
# Load cleaned dataset

listings = pd.read_csv("../data/listings_clean.csv")
```

```python
# View dataset information

print("Shape:", listings.shape)
listings.info()
```

```python
# Select features for modeling

features = [
    "accommodates",
    "bedrooms",
    "beds",
    "bathrooms",
    "review_scores_rating",
    "number_of_reviews",
    "availability_365",
    "room_type",
    "property_type"
]

model_data = listings[features + ["price_clean"]]

model_data.head()
```

```python
# Encode categorical variables

model_data = pd.get_dummies(
    model_data,
    columns=["room_type", "property_type"],
    drop_first=True
)

model_data.head()
```

```python
# Check for missing values

model_data.isnull().sum().sort_values(ascending=False).head(10)
```

```python
# Fill missing bathroom values

model_data["bathrooms"] = model_data["bathrooms"].fillna(
    model_data["bathrooms"].median()
)

model_data.isnull().sum().sort_values(ascending=False).head(10)
```

```python
# Save feature-engineered dataset

model_data.to_csv("../data/model_data.csv", index=False)
```
